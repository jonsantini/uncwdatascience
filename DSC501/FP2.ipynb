```python
def search(patterns,year=2019,ret=True):

    if isinstance(patterns,str):
        patterns = [patterns]
    
    sl = []
    for p in patterns:
        sl.append("re.search('"+p+"',x,re.I)")
    
    f = 'lambda x: '
    for s in sl:
        f += s+' and '
        
    d = pd.DataFrame(cd.search('acs5',year,'concept',eval(f[:-5])))
    d.columns = ['variable','concept','label']
    if len(d.concept.tail(1).values[0]) > 200:
        d = d.head(-1)
    u = d.concept.unique()
    print(len(u),' unique concept(s)')
    print(len(d),' rows in DataFrame')
    if ret:
        return u,d
```

```python
#3 

# Make a GeoDataFrame with the Massachusetts county shapes.
allcounties = gpd.read_file('/home/data/census/UScounties/')

# The state FIPS code for MA is 25
mass = allcounties[allcounties['STATEFP'] == '25']
mass.head()
```

```python
#3 continued

# Verify mass as a GeoDataFrame

mass.info()
```

```python
#3 continued

# Rename mass' COUNTYFP to fips so that there is a common column to merge on later on
mass.rename({'COUNTYFP': 'fips'}, axis=1, inplace=True)

# Verify the rename
mass.head()
```

```python
#3 continued

# Download from the census bureau the populations of the counties in Massachusetts.

# Recall that the population variable is 'B01003_001E'.

mass_pop = cd.download('acs5',2019,cd.censusgeo([('state','25'),('county','*')]),['B01003_001E'])
mass_pop.head()
```

```python
#3 continued

# I need to add in the FIPS codes

cg = mass_pop.index[0]
mass_pop['fips'] = [str(cg.geo[1][1]) for cg in mass_pop.index]
mass_pop.head()
```

```python
#3 continued

# Let's rename the columns

mass_pop.columns = ['population','fips']
mass_pop.head()
```

```python
#3 continued

# Need to confirm my fips codes are strings, not integers

mass_pop.info()
```

```python
#3 continued

# I need to merge my tables, mass & mass_pop. Make sure I merge mass first to retain the GeoDataFrame structure

mrge = pd.merge(mass, mass_pop)
mrge.head()
```

```python
#3 continued

# Verify mrge is a GeoDataFrame

mrge.info()
```

```python
#3 continued

mass_map = gv.Polygons(mrge,vdims=['population','NAME'])

mass_map = mass_map.opts(width=550,height=350,tools=['hover'],color='population',alpha=0.2,cmap='viridis_r')  * gv.tile_sources.CartoLight

mass_map
```

## Exercise

Read the file in to a DataFrame, parse the Date field to get a datetime object and make Date the index of the DataFrame. Some of the numeric columns have nulls so fill them with zeros.

It is possible to do all of this in one line with a call to read_csv().  It is also possible to first create the DataFrame, then make the Date column datetime, then make that the index and fill the nulls.  You may use either approach.

```python
#5

# Read in the file

crash = pd.read_csv('/home/data/airAccs.csv', index_col='Date', parse_dates=True, encoding = 'iso-8859-1')

# Fill up the nulls with zeroes

crash['Dead'] = crash['Dead'].fillna(0)
crash['Aboard'] = crash['Aboard'].fillna(0)
crash['Ground'] = crash['Ground'].fillna(0)

# Preview it

crash.head()
```

```python
#5 continued

# Verify it goes till 2014-03-22

crash.tail()
```

```python
#5 continued

# Verify DateTimeIndex & nulls are now non-null

crash.info()
```
