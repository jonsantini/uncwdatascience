You are going to find the number of arrests for each district in the district column of ar.


### Part 1

Use groupby() to make a DataFrame named d that has one column named Arrests and whose index is the districts in ar.district. Note that you want d to be a DataFrame, not a Series.

Note that ar contains many districts other than the six that are in wpd: D1,D2,D3,D4,D5,D6.  However, the bulk of the arrests are the six in wpd.

### Part 2

Redefine wpd to be the current wpd merged with d.  Here are the first few rows of wpd after the merge

<img src='../data/f1.2.png'>

```python
#4 Part 1

d = ar
d.head()
```

```python
#4 Part 1 continued

d = d.reset_index()
d
```

```python
#4 Part 1 continued

# Set district as the index

d=d.set_index('district')
d
```

```python
#4 Part 1 continued

# Find the number of arrests by district

# This is a series though

d2 = d.groupby("district")["case#"].count()
d2
```

```python
#4 Part 1 continued

# Convert series to DataFrame

d2 = d2.to_frame()
```

```python
#4 Part 1 continued

# Change the column name from the wrong column name ('case#') to the right column name ('arrests')

d2.rename(columns = {'case#':'arrests'}, inplace = True)
d2
```

```python
#4 Part 2 "pre-step"

# Reset the index so I can work with the district column

d3 = d2.reset_index()
d3
```

```python
#4 Part 2 "pre-step"

# Add a column named 'E911CODE' to d2 (so I can merge on a common column later)

# Drop the extra column

d3['E911CODE'] = d3['district'] 
d4 = d3.drop(['district'], axis=1)
d4.head()
```

```python
wpd.info()
```

```python
#4 Part 2

# Merge wpd and d4 on common column: E911CODE (which are the districts)
wpd = pd.merge(wpd, d4)
wpd
```

```python
# wpd is still a GeoDataFrame =)

wpd.info()
```
